```python
# ðŸ“Œ Product Prioritization Dashboard using RICE Scoring
# Author: Sowjanya Nadipena

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Define feature backlog
features = {
    "Feature": [
        "User Authentication",
        "Dark Mode",
        "Real-time Analytics",
        "Bug Reporting Tool",
        "Cloud Backup",
        "Multi-language Support"
    ],
    "Reach": [1000, 800, 600, 700, 500, 400],        # Number of users impacted
    "Impact": [3, 2, 3, 2, 3, 2],                    # 1=low, 2=medium, 3=high
    "Confidence": [0.9, 0.8, 0.7, 0.85, 0.95, 0.75], # 0-1 scale
    "Effort": [5, 2, 8, 4, 6, 3]                     # Story points (effort)
}

df = pd.DataFrame(features)

# Step 2: Calculate RICE score
df["RICE"] = (df["Reach"] * df["Impact"] * df["Confidence"]) / df["Effort"]

# Step 3: Sort features by RICE score
df_sorted = df.sort_values(by="RICE", ascending=False).reset_index(drop=True)

# Display ranked backlog
print("ðŸ“Œ Prioritized Feature Backlog (RICE Scoring):")
print(df_sorted[["Feature", "RICE"]])

# Step 4: Visualization
plt.figure(figsize=(10,6))
plt.barh(df_sorted["Feature"], df_sorted["RICE"], color="skyblue")
plt.xlabel("RICE Score")
plt.title("Feature Prioritization Dashboard (RICE Method)")
plt.gca().invert_yaxis()
plt.show()
```
